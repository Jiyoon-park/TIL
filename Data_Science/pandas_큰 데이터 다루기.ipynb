# 큰 DataFrame 살펴보기

```python
import pandas as pd
```

```python
laptops_df = pd.read_csv('laptops.csv')
laptops_df
```

### 데이터 위에 몇개만 보고 싶다면
* `df.head(int)` 를 사용한다.

```python
laptops_df.head(3)
```

```python
laptops_df.head(10)
```

### 데이터 아래 몇개만 보고 싶다면
* `df.tail(int)` 를 사용한다.

```python
laptops_df.tail(3)
```

### 데이터를 알고 싶다면
* 크기를 알고 싶다면 `df.shape` 를 찍어본다

```python
laptops_df.shape
```

* column들을 알고 싶다면 `df.colums` 를 찍어본다

```python
laptops_df.columns
```

* column들의 기본 정보를 알고 싶다면 `df.info()` 를 찍어보기
* row 수는 167 개인데, 167 개가 안되는 것들은 `비어있는 값`이라고 볼 수 있다.

```python
laptops_df.info()
```

* columns의 통계 값을 알고 싶어!
* `df.describe()`

```python
laptops_df.describe()
```

* 기준을 정해서 정렬하고 싶다면
* `df.sort_values(by='정렬기준')` (기본값 오름차순/ 내림차순으로 하고 싶다면 `ascending=False`)
* 원본이 변하는 것이 아니라 새로운 df가 생기는 것
* 기존 원본을 바꾸고 싶다면 `inplace=True`

```python
laptops_df.sort_values(by="price")
```

```python
laptops_df
```

```python
laptops_df.sort_values(by="price", ascending=False, inplace=True)
laptops_df
```
