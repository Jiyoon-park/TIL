## *새로운 값을 추가하거나 삭제하기*

```python
iphone_df = pd.read_csv('iphone.csv', index_col=0)
iphone_df
```

### 없는 값을 쓰면 새롭게 추가된다

```python
iphone_df.loc['iPhone 11'] = ['2019-11-21', '7', '7GB', 'iOS 13.0', 'Yes']
iphone_df
```

```python
iphone_df['제조사'] = 'PEACH'
iphone_df
```

### 값 삭제하기
* `.drop()`을 사용한다
* `axis='index'` 는 row를 , `axis='colums'`는 column을 뜻한다.
* `inplace=False` 를 쓰면 해당 값이 삭제된 새로운 데이터프레임이 생성되고, 원본은 변함이 없다.
* `inplace=True` 를 쓰면 원본에서 해당 값이 삭제된다.

```python
iphone_df.drop('iPhone 11', axis='index', inplace=False)
```

### 하지만 살아 있는 데이터?
* `inplace=False` 를 쓰면 기존의 데이터를 건들지 않겠다는 뜻이다.

```python
iphone_df
```

* `inplace=True` 를 쓰면 iphone_df 값 자체가 바뀌게 된다.

```python
iphone_df.drop('iPhone 11', axis='index', inplace=True)
iphone_df
```

```python
iphone_df
```

```python
iphone_df.drop('제조사', axis='columns', inplace=True)
iphone_df
```

### 여러줄 삭제할거야

```python
iphone_df.drop(['iPhone 7', 'iPhone 8', 'iPhone X'], axis='index', inplace=True)
iphone_df
```

## *index/ column 설정하기*

```python
liverpool_df = pd.read_csv('liverpool.csv', index_col=0)
liverpool_df
```

* `df.rename(columns={})`은 column의 이름을 바꿔 완전 새로운 df를 만든다.
* *원본은 변경되지 않음!*

```python
liverpool_df.rename(columns={'position': 'Position'})
```

```python
liverpool_df
```

* 원본을 바꾸고 싶다면, `inplace=True` 를 붙여준다.

```python
liverpool_df.rename(columns={'position': 'Position'}, inplace=True)
liverpool_df
```

```python
liverpool_df.rename(columns={'position': 'POSITION', 'born': 'BORN', 'number': 'NUMBER', 'nationality':'NALTIONALITY'})
```

```python
liverpool_df.index.name = 'player name'
liverpool_df
```

* 인덱스 이름을 선수 이름이 아닌 번호로 바꿔주고 싶다면 `df.set_index()`를 사용한다.

```python
liverpool_df.set_index('number')
```

* 하지만 이렇게 하면 선수 이름 정보가 다 없어지기 때문에, 선수 이름을 새로운 col으로 넣어주고 바꾸는게 좋다.

```python
liverpool_df['player name'] = liverpool_df.index
liverpool_df
```

```python
liverpool_df['player name'] = liverpool_df.index
liverpool_df
```

```python
liverpool_df.set_index('number', inplace=True)
liverpool_df
```
